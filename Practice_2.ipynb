```python
# find latitude and longtitude of a given location
```

```python
# import Nominatim class from the geocoders module of the geopy library
from geopy.geocoders import Nominatim

# create a Nominatim instance
geolocator = Nominatim(user_agent="geoapiExercises")

# location address
location_address = "27488 Stanford Avenue, North Dakota"

# lattitude and longitude coordinates for location address
coordinates = geolocator.geocode(location_address)

# print the location address and the cooordinates for it
print("address to find coordinates for: ", location_address)
print("\n")
print("latitude, longtitude: ", coordinates.latitude, coordinates.longitude)
```

```python
# search date from a given string
```

```python
# import arrow
import arrow

# get the date from a given string
date = arrow.get('David was born in 11 June 2003', 'DD MMMM YYYY')

# print the date
print("String: David was born in 11 June 2003")
print("\n")
print("Date from the string: ", date)
```

```python
# 3. Web Scraping
```

```python
# get the number of datasets currently listed on data.gov
```

```python
# import html class from lxml library
from lxml import html

# import requests library
import requests

# get the website link
link = requests.get('http://www.data.gov/')

# get the text from the link
text_from_link = html.fromstring(link.text)

# get datasets
datasets = text_from_link.cssselect('small a')[0]

# print the number of datasets
print(datasets.text)
```

```python
# 5. Matplotlib
```

```python
# import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# import the dataset
dataset = pd.read_csv('titanic.csv')

# view the dataset
dataset.head()
```

```python
# plot the numerical columns
dataset.plot()

# show the plot
plt.show()
```

```python
# plot the subplots
dataset.plot(subplots = True)

# show the plot
plt.show()
```

```python
# plot the subplots and fix the width size
dataset.plot(subplots = True, figsize = (15,12))

# show the plot
plt.show()
```

```python
# each subplot with it's on own x axis
dataset.plot(subplots = True, figsize = (15,12), sharex = False)

# show the plot
plt.show()
```

```python
# each subplot with the same y axis
dataset.plot(subplots = True, figsize = (15,12), sharex = False, sharey = True)

# show the plot
plt.show()
```

```python
# plot only the age column
dataset.age.plot(figsize = (15,12))

# show the plot
plt.show()
```

```python
# change the font size
dataset.age.plot(figsize = (15,12), fontsize = 13)

# show the plot
plt.show()
```

```python
# change the color
dataset.age.plot(figsize = (15,12), fontsize = 13, c = 'r')

# show the plot
plt.show()
```

```python
# change the linestyle
dataset.age.plot(figsize = (15,12), fontsize = 13, c = 'r', linestyle = ":")

# show the plot
plt.show()
```

```python
# add title to plot

dataset.age.plot(figsize = (15,12), fontsize = 13)
plt.title("Titanic - Ages", fontsize = 15)

# show the plot
plt.show()
```

```python
# add legend to plot

dataset.age.plot(figsize = (15,12), fontsize = 13)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)

# show the plot
plt.show()
```

```python
# add labels to axes

dataset.age.plot(figsize = (15,12), fontsize = 13)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger number", fontsize = 13)
plt.ylabel("Age", fontsize = 13)

# show the plot
plt.show()
```

```python
# add grid to the plot

dataset.age.plot(figsize = (15,12), fontsize = 13)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger number", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()

# show the plot
plt.show()
```

```python
# add xlim and ylim

dataset.age.plot(figsize = (15,12), fontsize = 13, xlim = (200, 400), ylim = (20, 40))
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger number", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()

# show the plot
plt.show()
```

```python
# add xticks and yticks

xticks = [x for x in range(0,901,50)]
yticks = [y for y in range(0,81,5)]

dataset.age.plot(figsize = (15,12), fontsize = 13, xlim = (0,900), ylim = (0,80), xticks = xticks, yticks = yticks)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger number", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()

# show the plot
plt.show()
```

```python
# rotate the ticks by 90 degrees
xticks = [x for x in range(0,901,50)]
yticks = [y for y in range(0,81,5)]

dataset.age.plot(figsize = (15,12), fontsize = 13, xlim = (0,900), ylim = (0,80), 
                 xticks = xticks, yticks = yticks, rot = 90)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger number", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()

# show the plot
plt.show()
```

```python
# rotate the ticks by 45 degrees
xticks = [x for x in range(0,901,50)]
yticks = [y for y in range(0,81,5)]

dataset.age.plot(figsize = (15,12), fontsize = 13, xlim = (0,900), ylim = (0,80), 
                 xticks = xticks, yticks = yticks, rot = 45)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger number", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()

# show the plot
plt.show()
```

```python
# import seaborn
import seaborn as sns

# plot with seaborn
plt.figure(figsize=(12,8))
sns.countplot(data = dataset, x = "sex")
plt.show()
```

```python
# separate the sex column based upon the pclass column

# plot with seaborn
plt.figure(figsize=(12,8))
sns.countplot(data = dataset, x = "sex", hue = "pclass")
plt.show()
```

```python
# set parameters

plt.figure(figsize=(12,8))
sns.set(font_scale=2)
sns.countplot(data = dataset, x = "sex", hue = "pclass")
plt.show()
```

```python
# heatmap


plt.figure(figsize=(12,8))
sns.heatmap(pd.crosstab(dataset.sex, dataset.pclass), annot = False, cmap = "Reds")
plt.show()
```
