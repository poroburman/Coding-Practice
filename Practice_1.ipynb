```python
# 1. GEOPY PACKAGE
```

```python
# write a python program to search the street address, name from a given location information
```

```python
# import Nominatim class from the geocoders module of the geopy library
from geopy.geocoders import Nominatim

# create the Nominatim instance
geolocator = Nominatim(user_agent="Address Exercise")

# street address to search the location of
location_address = "27488 Stanford Avenue, North Dakota"

# print the street address
print("Location address:", location_address)

# location of the street address
location = geolocator.geocode(location_address)

# print the location
print("\nStreet address and Street name: ")
print(location.address)
```

```python
# write a python program to get the current UTC datetime, local datetime, and datetime of a given location
```

```python
# import the arrow module
import arrow

# create an arrow instance for current datetime, local datetime, and specified local datetime
current_datetime = arrow.utcnow()
local_datetime = arrow.now()
specified_local_datetime = arrow.now('US/Mountain')

# print the datetimes
print("Current datetime: ", current_datetime)
print("\nCurrent local datetime: ", local_datetime)
print("\nCurrent ", specified_local_datetime)
```

```python
# write a python program to download and display the content of robot.txt for en.wikipedia.org
```

```python
# import requests module
import requests

# set an instance for requests
web_content = requests.get("https://en.wikipedia.org/robots.txt")

# extract the text from web_content
web_content_text = web_content.text

# print the text
print("robots.txt for https://en.wikipedia.org/robots.txt")
print("\n")
print(web_content_text)
```

```python
# convert string values to upper, lower cases, and find length of series
```

```python
# create the data series
data_series = pd.Series(['X', 'Y', 'Z', 'Aaba', 'Baca', np.nan, 'CABA', None, 'bird', 'horse', 'dog'])

# print the data_series
print(data_series)

# print the upper cases of data_series
print("\n")
print(data_series.str.upper())

# print the lower cases of data_series
print("\n")
print(data_series.str.lower())

# print length of series
print("\n")
print(data_series.str.len())
```

```python
# merge two dataframes
```

```python
# dataframe 1
dataframe_1 = pd.DataFrame({
    'student_id':['S1','S2','S3','S4','S5'],
    'name':['Daniella Fenton','Ryder Storey','Bryce Jensen','Ed Bernal','Kwame Morin'],
    'marks':[200,210,190,222,199]})

# dataframe 2
dataframe_2 = pd.DataFrame({
    'student_id':['S4','S5','S6','S7','S8'],
    'name':['Scarlette Fisher','Carla Williamson','Dante Morse','Kaiser William','Madeeha Preston'],
    'marks':[201,200,198,219,201]})

# merge the two dataframes
merged_dataframe = pd.concat([dataframe_1, dataframe_2])

# print the individual dataframes and the merged dataframe
print(dataframe_1)
print("\n")
print(dataframe_2)
print("\n")
print(merged_dataframe)
```

```python
# draw a line with suitable labels
```

```python
# import matplotlib
import matplotlib.pyplot as plt

# set the data
x = range(1,50)
y = [value*3 for value in x]

# plot the data
plt.plot(x,y)

# set axes labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# set a title
plt.title('Draw a line')

# display the figure
plt.show()
```
